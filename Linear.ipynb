```python
import pandas as pd
poverty = pd.read_excel("Family Poverty.xlsx")
poverty.dropna(subset = ["Average Tax Rate"], inplace=True)
poverty = poverty[poverty["Year"] < 2008]
```

```python
poverty
```

```python
y = poverty["Poverty rate for families"].to_frame()
```

```python
x = poverty["Average Tax Rate"].to_frame()
x
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x, y)
```

```python
reg.score(x, y)
```

```python
reg.coef_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y, 'o')
plt.plot(x, reg.coef_* x + reg.intercept_)
```

```python
unemployment = pd.read_csv("US_unemployment.csv")
```

```python
ally = pd.read_csv("US_ally.csv")
```

```python
df = pd.merge(unemployment, ally,  how='left', left_on=['Year'], right_on = ['year'])
```

```python
df = df[df["year"] < 2008]
df
```

```python
X = df["c_avg"].to_frame()
Y = df["rate"].to_frame()
```

```python
reg = LinearRegression().fit(X, Y)
reg.score(X,Y)
```

```python
plt.plot(X, Y, 'o')
plt.plot(X, reg.coef_* X + reg.intercept_)
```
